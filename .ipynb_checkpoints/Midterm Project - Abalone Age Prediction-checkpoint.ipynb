```python
df_abalone = pd.read_csv("abalone.csv")
df_abalone
```

## Lower Case all the columns

```python
def lower_case_column_names(df_abalone):
    df_abalone.columns=[i.lower() for i in df_abalone.columns]
    return df_abalone
```

```python
df_abalone = lower_case_column_names(df_abalone)
df_abalone
```
